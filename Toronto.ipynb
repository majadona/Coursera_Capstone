<h1>Web scraping and importnig libraries</h1>

```python
import pandas as pd 

import requests 

from bs4 import BeautifulSoup 

 

req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M") 

soup = BeautifulSoup(req.content,'lxml') 

table = soup.find_all('table')[0]  

df = pd.read_html(str(table)) 

 

neighborhood=pd.DataFrame(df[0])
```

```python
neighborhood
```

<h1> Joining Postal Codes and Geospatial Data</h1>

```python
df_postcode = neighborhood.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()

df_postcode.head()
```

```python
df_postcode
```

```python
locgeo_df = pd.read_csv('https://cocl.us/Geospatial_data', index_col='Postal Code')

toronto_data = df_postcode.join(locgeo_df, on='Postal Code')
```
